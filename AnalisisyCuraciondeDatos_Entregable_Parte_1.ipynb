#### 1.3. Implementar consultas en SQL que respondan con la siguiente información:
 

*   Cantidad de registros totales por ciudad.
*   Cantidad de registros totales por barrio y ciudad.


  



Cantidad de registros totales por ciudad.


```python
query_1 = '''

SELECT city, COUNT(*) as xxx
FROM airbnb
GROUP BY city
'''

with engine.connect() as connection:
    result = connection.execute(query_1)
    for row in result:
       print(row)
    
```

Cantidad de registros totales por barrio y ciudad.

```python
query_2 = '''
SELECT city, neighborhood, COUNT(*)
FROM airbnb
GROUP BY city, neighborhood
'''

with engine.connect() as connection:
    result = connection.execute(query_2)
    for row in result:
       print(row)
```

####1.4. Combinar los datasets de ambas tablas ingestadas utilizando el comando JOIN de SQL  para obtener un resultado similar a lo realizado con Pandas en clase. 

```python
#Genero la consulta JOIN de SQL:

query = '''

SELECT *
FROM melb LEFT JOIN airbnb ON Postcode = zipcode
LIMIT 10
'''
#A fin de que este bloque se ejecute lo más rápidamente posible se incorporó el statement LIMIT 10 al final de la última consulta. 
#Para lograr un LEFT JOIN completo de las dos tablas es necesario quitar ese statement.

#Ejecuto la consulta:
with engine.connect() as connection:
    result = connection.execute(query)
    for row in result:
      print(row)
```

```python
#Se estandarizan los datos:

airbnb_df['zipcode'] = pd.to_numeric(airbnb_df.zipcode, errors='coerce')


#Se obtiene la frecuencia de los zipcodes incluyendo solo los que tienen una
#frecuencia absoluta minima de 30:

airbnb_df.zipcode.value_counts().loc[lambda x : x>20]
```

```python
#Se suma la cantidad de zipcode registrados en los datos:

print('Cantidad de registros totales: ' + str(airbnb_df.zipcode.value_counts().sum()))



#Se suma la cantidad de zipcode con la frecuencia de 20:

print('Cantidad de zipcodes con mas de 20 registros: ' + str(airbnb_df.zipcode.value_counts().loc[lambda x : x>20].sum()))

print('\n')

total = airbnb_df.zipcode.value_counts().sum()
filtrado = airbnb_df.zipcode.value_counts().loc[lambda x : x>20].sum()


#Se obtiene el porcentaje del total:

print('Porcentaje de registros con el que se va a trabajar del total: ' + str(round(filtrado/total,2)*100) + '%' )

```

```python
#Se genera una lista con los valores con una frecuencia mayor a 20:

filtrado = airbnb_df.zipcode.value_counts().loc[lambda x : x>20].index.to_list()


#Se genera un nuevo dataframe con esta condicion filtrada:

airbnb_filtered_df = airbnb_df[airbnb_df['zipcode'].isin(filtrado)]


#Se comprueba que la cantidad filtrada sea igual a la obtenida anteriormente:

len(airbnb_filtered_df)
```

```python
#Dataframe de airbnb unicamente con columnas zipcode y price.
prices_df = airbnb_filtered_df.groupby('zipcode', as_index=False).mean()[['zipcode','price']]
prices_df
```

```python
#Se convierten la columna zipcode a numerica.
prices_df['zipcode'] = pd.to_numeric(prices_df.zipcode, errors='coerce')
```

```python
#Se realiza un merge entre melb_df y prices_df.
merge_df = melb_df_selected_cols.merge(prices_df, how = 'left', left_on = 'Postcode', right_on = 'zipcode')
merge_df
```

```python
#Se contea la cantidad de precios mayores a cero.
merge_df[merge_df.price > 0][['price']].count()
```

```python
#Se visualiza la cantidad de nulos por columna.
merge_df.isna().sum()
```

```python
#Se realiza una interseccion entre los valores unicos de zipcode de airbnb y los valores unicos de postcode de melbourne.
intersection = np.intersect1d(
    airbnb_filtered_df.zipcode.values, melb_df.Postcode.values, assume_unique=False)
print("Airbnb unique zipcodes", len(airbnb_filtered_df.zipcode.unique()))
print("Sales unique zipcodes", len(melb_df.Postcode.unique()))
print("Common zipcodes", len(intersection))
```
